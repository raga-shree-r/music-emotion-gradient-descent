# Music Emotion Prediction using Gradient Descent

## Introduction
Music plays a significant role in influencing human emotions. However, most music recommendation systems rely on user behavior rather than directly understanding emotional responses.

This project aims to predict the emotional response to music using features such as genre and user rating. Gradient descent is used as an optimization technique to minimize prediction error and improve model performance.

## Problem Statement

The goal of this project is to develop a model that can predict the emotional response associated with music based on its characteristics.

The aim is to use gradient descent to minimize the error between predicted and actual emotions, thereby improving the accuracy of the model.

## Data Loading and Exploration

```python
import pandas as pd

df = pd.read_csv("musicemotions_dataset_v2.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Data Preprocessing

Irrelevant features such as file names and identifiers were removed. Only meaningful features were retained for modeling.

```python
df.isnull().sum()
```

```python
df = df[["Genre", "Rating", "Feeling"]]
```

## Feature Encoding and Scaling

Categorical variables were converted into numerical form, and feature scaling was applied to ensure efficient gradient descent optimization.

```python
df["Genre"] = df["Genre"].astype("category").cat.codes
```

```python
df["Feeling"] = df["Feeling"].astype("category").cat.codes
```

```python
X = df[["Genre", "Rating"]]
y = df["Feeling"]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## Basic Model using Gradient Descent

A simple linear model is implemented to predict emotions using gradient descent optimization.

```python
import numpy as np

# Convert to numpy
X = np.array(X_scaled)
y = np.array(y)

# Initialize weights and bias
w = np.zeros(X.shape[1])
b = 0

# Hyperparameters
learning_rate = 0.01
epochs = 1000

losses = []
```

```python
for i in range(epochs):

    # Predictions
    y_pred = np.dot(X, w) + b

    # Loss
    loss = np.mean((y - y_pred) ** 2)
    losses.append(loss)

    # Gradients
    dw = (-2/len(y)) * np.dot(X.T, (y - y_pred))
    db = (-2/len(y)) * np.sum(y - y_pred)

    # Update
    w = w - learning_rate * dw
    b = b - learning_rate * db
```

## Loss vs Iterations (Basic Model)

The graph shows how the loss decreases over iterations, indicating that gradient descent is successfully minimizing the error.

```python
print(losses[:5])
print(losses[-5:])
```

```python
import matplotlib.pyplot as plt

plt.plot(losses)
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.title("Loss vs Iterations")
plt.show()
```

The loss decreases rapidly in the initial iterations and then stabilizes, showing that the model converges to an optimal solution.

```python
y_pred = np.dot(X, w) + b

import matplotlib.pyplot as plt

plt.scatter(y, y_pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted")
plt.show()
```
